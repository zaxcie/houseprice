```python
%load_ext autoreload
%autoreload 2
%reload_ext autoreload

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from src.data.process import HousePriceData

data = HousePriceData("/Users/kforest/workspace/houseprice/data/raw/train.csv",
                      "/Users/kforest/workspace/houseprice/data/raw/test.csv")
```

```python
from src.configs import *
from src.features.transform import transform_quality_var_into_ordinal
from src.analysis import linear_relationship_between_target_and_ordinal_var,\
    multilinear_relationship_between_target_and_ordinal_var
from src.features.impute import impute_var_mean
import warnings
import time
import copy


warnings.filterwarnings('ignore')

start_time = time.time()

temp_workset, ordinal_feature_name, _ = transform_quality_var_into_ordinal(data.workset)
data.update_workset(temp_workset)

train = copy.deepcopy(data.trainset)

for i in CONTINUOUS_COL + DISCRETE_COL + ordinal_feature_name:
    train[i] = impute_var_mean(train[i])
    
multilinear_relationship_between_target_and_ordinal_var(train[CONTINUOUS_COL + DISCRETE_COL + ordinal_feature_name],
                                                        data.target)
```

```python
start_time = time.time()

warnings.filterwarnings('ignore')

temp_workset, ordinal_feature_name, _ = transform_quality_var_into_ordinal(data.workset)
data.update_workset(temp_workset)

r2_collection = dict()
pearson_collection = dict()

for i in CONTINUOUS_COL + DISCRETE_COL + ordinal_feature_name:
    print(i)
    try:
        data_subset_index = data.trainset[(data.trainset[i] != 'NO') & (data.trainset[i] != 'MISSING')].index
    except Exception as e:
        data_subset_index = data.trainset.index
        
    ord_var = np.array(data.trainset.iloc[data_subset_index][i])
    r2, pearson = linear_relationship_between_target_and_ordinal_var(ord_var,
                                                            data.target.iloc[data_subset_index],
                                                            i)
    r2_collection[i] = r2
    pearson_collection[i] = pearson

print(sorted(r2_collection.items(), key=lambda x: x[1], reverse=True))
print(sorted(pearson_collection.items(), key=lambda x: x[1], reverse=True))   
print("--- %s seconds ---" % (time.time() - start_time))
```
